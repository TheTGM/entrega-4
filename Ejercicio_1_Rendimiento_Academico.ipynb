## Punto 1 - Descripción del Problema

El objetivo de este ejercicio es construir un modelo predictivo capaz de clasificar el rendimiento académico de los estudiantes en tres categorías:

- Bajo
- Medio
- Alto

A partir de variables como:
- Hábitos de estudio
- Acceso a internet
- Relaciones familiares
- Tiempo de estudio
- Educación de los padres
- Consumo de alcohol
- Calificaciones anteriores

El problema corresponde a una clasificación supervisada multiclase.
